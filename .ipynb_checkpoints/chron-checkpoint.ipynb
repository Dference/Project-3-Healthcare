```python
# Import dependencies
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read from CSV file into DataFrame -
df = pd.read_csv('Updated_Deaths_Sheet.csv')
```

```python
# Print DataFrame to check data
print(df.head())
```

```python
# Use Seaborn to create a scatter plot of Chron_Cause vs Non_Chron_Causes
plt.figure(figsize=(10, 10))
sns.scatterplot(data=df, x='chron_causes', y='jurisdiction')

# Add titles and labels
plt.title('Chronic cases by Jurisdiction')
plt.xlabel('Chron_Cause')
plt.ylabel('Jurisdiction')

# Show plot
plt.show()
```
